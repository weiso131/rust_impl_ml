```python
import random


def gernerate_data(a_num, b_num):


# 設定群的中心點
    cluster_1_center = [2, 2]
    cluster_2_center = [8, 8]

    # 生成 15 筆資料分布在第一群
    cluster_1 = [[round(random.gauss(cluster_1_center[0], 1), 3),
                round(random.gauss(cluster_1_center[1], 1), 3)] for _ in range(a_num)]

    # 生成 15 筆資料分布在第二群
    cluster_2 = [[round(random.gauss(cluster_2_center[0], 1), 3),
                round(random.gauss(cluster_2_center[1], 1), 3)] for _ in range(b_num)]

    # 合併兩個群的資料
    data = cluster_1 + cluster_2

    # 將資料展平為一維 list
    

    return data

data = gernerate_data(15, 15)
data_1d = [value for sublist in data for value in sublist]
print(data_1d)
```

```python
import matplotlib.pyplot as plt
# 將資料轉換為 x 和 y 的形式
x_values = [point[0] for point in data]
y_values = [point[1] for point in data]

# 繪製散點圖
plt.scatter(x_values, y_values, c='blue', label='Data Points')
plt.title('Scatter Plot of Random Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```
